# Test Result

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('result.csv',sep=';')
```

```python
data.tail()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['timestamp'], data['profit_and_loss'])
plt.title('Profit and Loss Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Profit/Loss')
plt.grid(True)
plt.show()
```

```python
kelp=data[data['product']=='KELP']

plt.figure(figsize=(12, 6))
plt.plot(kelp['timestamp'], kelp['bid_price_1'], label='Bid Price 1')
plt.plot(kelp['timestamp'], kelp['bid_price_2'], label='Bid Price 2')
plt.plot(kelp['timestamp'], kelp['bid_price_3'], label='Bid Price 3')
plt.title('KELP Bid Prices Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(kelp['timestamp'], kelp['ask_price_2'], label='Bid Price 2')
plt.plot(kelp['timestamp'], kelp['ask_price_1'], label='Bid Price 1')
plt.plot(kelp['timestamp'], kelp['ask_price_3'], label='Bid Price 3')
plt.title('KELP Bid Prices Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
# plt.plot(kelp['timestamp'], kelp['ask_price_2']-kelp['bid_price_2'], label='Bid Price 2')
plt.scatter(kelp['timestamp'], kelp['ask_price_1']-kelp['bid_price_1'], label='Bid Price 1')
# plt.plot(kelp['timestamp'], kelp['ask_price_3']-kelp['bid_price_3'], label='Bid Price 3')
plt.title('Difference of Bid and ask ')
plt.xlabel('Timestamp')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
rain=data[data['product']=='RAINFOREST_RESIN']

plt.figure(figsize=(12, 6))
# plt.plot(rain['timestamp'], rain['ask_price_2']-rain['bid_price_2'], label='Bid Price 2')
plt.plot(rain['timestamp'], rain['ask_price_1']-rain['bid_price_1'], label='Bid Price 1')
# plt.plot(rain['timestamp'], rain['ask_price_3']-rain['bid_price_3'], label='Bid Price 3')
plt.title('Difference of Bid and ask ')
plt.xlabel('Timestamp')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))

kelp_spread = kelp['ask_price_1'] - kelp['bid_price_1']
rain_spread = rain['ask_price_1'] - rain['bid_price_1']


plt.subplot(1, 2, 1)
plt.hist(kelp_spread, bins=30, alpha=0.7)
plt.title('KELP Bid-Ask Spread Distribution')
plt.xlabel('Spread')
plt.ylabel('Frequency')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.hist(rain_spread, bins=30, alpha=0.7)
plt.title('RAINFOREST_RESIN Bid-Ask Spread Distribution')
plt.xlabel('Spread')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()
```
